```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Quantificação de Metanol:**

```python
file_path = 'metanol_v1.xlsx'
sheet_name = 'todos'
df = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
df.head()
```

```python
ax = df.plot(kind='bar', x='Amostra', y='[CH3OH] (mol.L-1) (média)', yerr='Desvio padrão', capsize=5, legend=False)

ax.set_ylabel('[CH3OH] (mol.L-1)')
ax.set_xlabel('')
ax.set_title('Todos')

plt.show()
plt.savefig('metanol_todos.png')
```

```python
df_hcl = df[df['Tratamento Químico'] == 'HCl']

ax = df_hcl.plot(kind='bar', x='Amostra', y='[CH3OH] (mol.L-1) (média)', yerr='Desvio padrão', capsize=5, legend=False, color='purple')

ax.set_ylabel('[CH3OH] (mol.L-1)')
ax.set_xlabel('')
ax.set_title('HCl')

plt.show()
plt.savefig('metanol_hcl.png')
```

```python
df_naoh = df[df['Tratamento Químico'] == 'NaOH']

ax = df_naoh.plot(kind='bar', x='Amostra', y='[CH3OH] (mol.L-1) (média)', yerr='Desvio padrão', capsize=5, legend=False, color='teal')

ax.set_ylabel('[CH3OH] (mol.L-1)')
ax.set_xlabel('')
ax.set_title('NaOH')

plt.show()
plt.savefig('metanol_naoh.png')
```

## **Perda de Massa:**

```python
file_path = 'perda_de_massa_v1.xlsx'
sheet_name = 'etapa1_medias'
df_et1 = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
df_et1_hcl = df_et1[df_et1['Tratamento Químico'] == 'HCl 2M']

ax = df_et1_hcl.plot(kind='bar', x='Amostra', y='Perda de massa_raw (%)', yerr='desvio_raw', capsize=5, legend=False, color='red')

ax.set_ylabel('Perda de massa_raw (%)')
ax.set_xlabel('')
ax.set_title('Perda de Massa - Degradação Química (HCl 2M) - Sem Exclusões')

plt.show()
plt.savefig('massa_et1_hcl_semexc.png')
```
